```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
def MA(q,path,col):
    data = pd.read_csv(path)
    data=data[col].to_frame()
    for i in range(1,q+1):
        data['Shifted_values_%d' % i ] = data[col].shift(i)

    data_2 = data.dropna()
    X_train = data_2.iloc[:,1:].values.reshape(-1,q)
    y_train = data_2.iloc[:,0].values.reshape(-1,1)

    lr = LinearRegression()
    lr.fit(X_train,y_train)

    theta  = lr.coef_.T
    intercept = lr.intercept_
    data_2['Predicted'] = X_train.dot(lr.coef_.T) + lr.intercept_

    #data_2[[col,'Predicted']].plot()

    
    RMSE = np.sqrt(mean_squared_error(data_2[col], data_2['Predicted']))

    print("The RMSE is :", RMSE,", Value of q : ",q)
    return data_2['Predicted']
```

```python
MA(4, 'state_diff/California.csv', 'confirmed_diff')
```

```python
for state in stateList:
    predicted_death=MA(4, 'state_diff/{}.csv'.format(state), 'death_diff')
    predicted_confirmed=MA(4, 'state_diff/{}.csv'.format(state), 'confirmed_diff')
```
